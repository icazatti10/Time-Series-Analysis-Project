```python
# import libraries
from data_registry import datasets
import pandas as pd
import numpy as np
from statsmodels.gam.api import GLMGam
from statsmodels.gam.api import BSplines
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA

# utils functions
from utils.mainutil import to_ts
from utils.m1util import get_spline

# internal plot functions
from plots.ts_plot import ts_plot
from plots.residual_plot import residual_plot
```

```python
# import weather 10 minute data
filename = datasets.get("weather_10_min")
df = pd.read_csv(filename)

# print(df.columns) - feature to study temperature
df = df[["date", "T (degC)"]].copy()
df.index = pd.to_datetime(df["date"])
df = df.drop(columns="date")

# convert from celsius to farenheit
df["ts"] = df["T (degC)"].apply(lambda x: x * 1.8 + 32)
df = df.drop("T (degC)", axis=1)

fig1 = ts_plot(x=df.index, y=df["ts"], title="Temperature (F) - 10 minutes", xlabel="Date", ylabel="in")
```

### Estimate Temperature Trend

```python
ts = to_ts(df["ts"])

# Estimate trend using GAM
ts["doy"] = np.arange(len(ts))
splines = BSplines(ts[["pts", "doy"]],df=[20, 10],degree=[3, 3])
gam_model = GLMGam.from_formula("ts ~ 1",data=ts,smoother=splines,alpha=[10, 10],family=sm.families.Gaussian()).fit()

# Fit + Residual values
ts["gam_fit"] = gam_model.fittedvalues
ts["gam_res"] = ts["ts"] - ts["gam_fit"]

# Data + Trends Plot
fig = plt.figure(figsize=(12, 4))
plt.plot(ts.index, ts["ts"])
plt.plot(ts.index, ts["gam_fit"])
plt.title("10 Mintue Temperature Data")
plt.ylabel("Temperature [F]")
plt.xlabel("Date")
plt.grid(True)
plt.show()

# Residual Plots
plt.figure(figsize=(12,4))
plt.plot(ts["gam_res"] )
plt.xlabel("Date")
plt.ylabel("Residual")
plt.grid(True)
plt.show()

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

plot_acf(ts["gam_res"] , lags=80, ax=axs[0])
axs[0].set_title("ACF of Residual")

plot_pacf(ts["gam_res"] , lags=80, ax=axs[1], method="ywm")
axs[1].set_title("PACF of Residual")

plt.tight_layout()
plt.show()
```

```python
resid = ts["gam_res"].copy()
resid.index = pd.to_datetime(resid.index)
resid = resid.sort_index()
resid = resid[~ts.index.duplicated(keep="first")]
resid = resid.asfreq("10min")
resid = resid.astype(float)

resid = resid.astype(float)

model_ar1 = ARIMA(resid, order=(1,0,0)).fit()
model_ar2 = ARIMA(resid, order=(2,0,0)).fit()
model_arma11 = ARIMA(resid, order=(1,0,1)).fit()
model_arma31 = ARIMA(resid, order=(3,0,1)).fit()
model_arma41 = ARIMA(resid, order=(4,0,1)).fit()
model_arma42 = ARIMA(resid, order=(4,0,2)).fit()

print("AR(1) AIC:", model_ar1.aic)
print("AR(2) AIC:", model_ar2.aic)
print("ARMA(1,1) AIC:", model_arma11.aic)
print("ARMA(3,1) AIC:", model_arma31.aic)
print("ARMA(4,1) AIC:", model_arma41.aic)
print("ARMA(4,2) AIC:", model_arma42.aic)
```

```python
best_model = model_arma41
ts["arma_res"] = best_model.resid


# Residual Plots
plt.figure(figsize=(12,4))
plt.plot(ts["arma_res"])
plt.xlabel("Date")
plt.ylabel("Residual")
plt.grid(True)
plt.show()

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

plot_acf(ts["arma_res"] , lags=80, ax=axs[0])
axs[0].set_title("ACF of Residual")

plot_pacf(ts["arma_res"] , lags=80, ax=axs[1], method="ywm")
axs[1].set_title("PACF of Residual")

plt.tight_layout()
plt.show()
```
